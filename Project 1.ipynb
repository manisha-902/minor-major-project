# PART : A Data processing

# STEP 1: Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# STEP2: Import dataset

```python
dataset=pd.read_csv('investment_data.csv')
```

```python
dataset
```

# STEP 3: To create feature matrix(X) and dependent variable vector(Y)

```python
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
X
```

```python
Y
```

# STEP 4: Encoding

# Feature Matrix using OneHotEncoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
X
```

# STEP 5: Replacing the missing data

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[::])
(X[::])=imputer.transform(X[::])
```

```python
X
```

# STEP 6: Spliting the data set into training data set and testing data set

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
Xtrain
```

# STEP 7: Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Xtrain=sc.fit_transform(Xtrain)
Xtest=sc.fit_transform(Xtest)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
Y=np.array(le.fit_transform(Y))
```

```python
Y
```

# STEP B: Build classification mode

# STEP1: Training the linear model

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(Xtrain,Ytrain)
```

# STEP 2: Testing the linear model

```python
Yestimated=LR.predict(Xtest)
```
